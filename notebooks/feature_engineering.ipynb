```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import (
    LabelEncoder,
    StandardScaler
)
```

```python
df = pd.read_csv(
    "../data/processed/master_data.csv"
)

df.head()
```

```python
df.fillna(
    df.median(
        numeric_only=True
    ),
    inplace=True
)
```

```python
encoder = LabelEncoder()

categorical_columns = [
    "gender",
    "diagnosis",
    "insurance_type"
]

for col in categorical_columns:

    df[col] = encoder.fit_transform(
        df[col]
    )
```

```python
scaler = StandardScaler()

numeric_cols = [
    "age",
    "income",
    "claim_amount"
]

df[numeric_cols] = scaler.fit_transform(
    df[numeric_cols]
)
```

```python
df.to_csv(
    "../data/features/ml_features.csv",
    index=False
)
```
