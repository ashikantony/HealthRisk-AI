```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sim_df = pd.read_csv(
    "../data/simulation/simulation_results.csv"
)

sim_df.head()
```

```python
sim_df.groupby(
    "shock"
)["portfolio_value"] \
.mean()
```

```python
sim_df.groupby(
    "round"
)["portfolio_value"] \
.mean() \
.plot(
    figsize=(12, 6)
)

plt.title(
    "Portfolio Value by Round"
)

plt.show()
```

```python
sim_df.groupby(
    "player"
)["score"] \
.max() \
.sort_values(
    ascending=False
) \
.head(20)
```
