```python
from sklearn.ensemble import (
    RandomForestClassifier
)
```

```python
df = pd.read_csv(
    "../data/features/pharma_features.csv"
)
```

```python
X = df.drop(
    columns=["drug_success"]
)

y = df["drug_success"]
```

```python
model = RandomForestClassifier(
    n_estimators=500,
    random_state=42
)

model.fit(X, y)
```

```python
importance = pd.DataFrame({

    "feature":
        X.columns,

    "importance":
        model.feature_importances_
})

importance.sort_values(
    "importance",
    ascending=False
).head(20)
```
