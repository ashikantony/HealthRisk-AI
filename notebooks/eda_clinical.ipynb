```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

sns.set_style("whitegrid")
```

```python
clinical_df = pd.read_csv(
    "../data/processed/clinical_data.csv"
)

clinical_df.head()
```

```python
clinical_df.info()
clinical_df.describe()
```

```python
missing = (
    clinical_df.isnull()
    .sum()
    .sort_values(ascending=False)
)

missing
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(
    clinical_df["age"],
    bins=30,
    kde=True
)

plt.title("Age Distribution")
plt.show()
```

```python
clinical_df["diagnosis"] \
    .value_counts() \
    .head(20) \
    .plot(
        kind="bar",
        figsize=(12, 6)
    )

plt.title("Top Diagnoses")
plt.show()
```

```python
mortality = (
    clinical_df
    .groupby("diagnosis")
    ["mortality"]
    .mean()
)

mortality.sort_values(
    ascending=False
).head(20)
```

```python
corr = (
    clinical_df
    .select_dtypes(
        include=np.number
    )
    .corr()
)

plt.figure(figsize=(14, 10))

sns.heatmap(
    corr,
    cmap="coolwarm"
)

plt.show()
```
