```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
finance_df = pd.read_csv(
    "../data/processed/financial_data.csv"
)

finance_df.head()
```

```python
finance_df.describe()
```

```python
finance_df["claim_amount"] \
    .hist(
        bins=40,
        figsize=(10, 6)
    )

plt.title(
    "Claim Amount Distribution"
)

plt.show()
```

```python
sns.boxplot(
    data=finance_df,
    x="insurance_type",
    y="claim_amount"
)

plt.xticks(rotation=45)

plt.show()
```

```python
finance_df.groupby(
    "hospital"
)["revenue"] \
.sum() \
.sort_values(
    ascending=False
) \
.head(20)
```
