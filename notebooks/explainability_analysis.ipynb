```python
import pandas as pd
import shap

from xgboost import (
    XGBClassifier
)
```

```python
df = pd.read_csv(
    "../data/features/ml_features.csv"
)

X = df.drop(
    columns=["target"]
)

y = df["target"]
```

```python
model = XGBClassifier()

model.fit(X, y)
```

```python
explainer = shap.Explainer(
    model,
    X
)

shap_values = explainer(X)
```

```python
shap.summary_plot(
    shap_values,
    X
)
```

```python
shap.plots.waterfall(
    shap_values[0]
)
```

```python
importance = pd.DataFrame({

    "feature":
        X.columns,

    "importance":
        abs(
            shap_values.values
        ).mean(axis=0)
})

importance.sort_values(
    "importance",
    ascending=False
)
```
